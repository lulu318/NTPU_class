```python
import pandas as pd

# 讀取 CSV 檔案
df = pd.read_csv('D:/class/Regression_Analysis/data/Dengue_Daily.csv')
df['發病日'] = pd.to_datetime(df['發病日'], format='%Y/%m/%d')
filtered_df = df[(df['發病日'] >= '2023-01-01') & (df['發病日'] <= '2023-12-31') & (df['是否境外移入'] == '否')]
# filtered_df.shape
# filtered_df.to_csv('D:/class/Regression_Analysis/data/filtered_data.csv', index=False, encoding='utf-8-sig')
grouped = filtered_df.groupby(['發病日', '居住縣市', '居住鄉鎮'])['確定病例數'].sum().reset_index()
# grouped.to_csv('D:/class/Regression_Analysis/data/登革熱2022-2023年每日確定病例分組統計.csv', index=False, encoding='utf-8-sig')
kaohsiung_dengue_data = grouped[grouped['居住縣市'] == '高雄市']

print(grouped.describe())
print(grouped.shape)

kaohsiung_dengue_data = kaohsiung_dengue_data.rename(columns={'居住鄉鎮': '測站', '發病日': '日期'})
```

```python
import glob
import os
import pandas as pd

folder_path = 'D:/class/Regression_Analysis/data/高雄空品區_2023'
file_paths = glob.glob(os.path.join(folder_path, '*.csv'))


all_data = []

# 逐一處理每個檔案
for file_path in file_paths:
    dff = pd.read_csv(file_path)
    dff = dff.drop(columns=['Unnamed: 27'], errors='ignore')

    dff['測站'] = dff['測站'] + '區'
    hour_columns = dff.columns[3:27]
    dff_ = dff.melt(id_vars=['測站', '日期', '測項'], value_vars=hour_columns, var_name='小時', value_name='值')
    dff_['值'] = pd.to_numeric(dff_['值'], errors='coerce')
    # print(dff_.head())
    dff_['日期'] = pd.to_datetime(dff_['日期']).dt.strftime('%Y-%m-%d')

    df_pivot = pd.pivot_table(
        dff_,
        index=['測站', '日期'],
        columns='測項',  # 直接指定 '測項' 欄位
        values='值',
        aggfunc=['min', 'max', 'mean', 'sum']
    ).reset_index()
    df_pivot.columns = ['_'.join(col).strip() if col[1] else col[0] for col in df_pivot.columns.values]
    all_data.append(df_pivot)
final_data = pd.concat(all_data, ignore_index=True)
```

```python
kaohsiung_dengue_data['日期'] = pd.to_datetime(kaohsiung_dengue_data['日期'])
final_data['日期'] = pd.to_datetime(final_data['日期'])
merged_data = pd.merge(final_data, kaohsiung_dengue_data, on=['日期', '測站'], how='left')

merged_data['確定病例數'] = merged_data['確定病例數'].fillna(0)
north = ['三民區', '鼓山區', '左營區', '楠梓區', '仁武區', '橋頭區', '美濃區']
south = ['鹽埕區', '新興區', '前金區', '苓雅區', '前鎮區', '小港區', '大寮區', '復興區', '林園區', '鳳山區']
merged_data['area'] = merged_data['測站'].apply(lambda x: 1 if x in north else 2 if x in south else 0 )

merged_data = merged_data.rename(columns={'測站': 'station', '日期': 'date', '確定病例數': 'item', '有無確定病例數':'item_bool'})
merged_data = merged_data.drop(columns=['居住縣市'])
merged_data.to_csv('C:/Users/USER/regression/final.csv', index=False, encoding='utf-8-sig')
```

```python
merged_data_select = merged_data[merged_data['item'] != 0]
merged_data_select['date_n'] = pd.to_datetime(merged_data_select['date'])
merged_data_select['year'] = merged_data_select['date_n'].dt.year
merged_data_select['month'] = merged_data_select['date_n'].dt.month
def determine_season(month):
    if month in [3, 4, 5]:
        return '春季'
    elif month in [6, 7, 8]:
        return '夏季'
    elif month in [9, 10, 11]:
        return '秋季'
    else:
        return '冬季'

merged_data_select['season'] = merged_data_select['month'].apply(determine_season)
merged_data_select = merged_data_select.drop(columns=['date_n'])

merged_data_select = merged_data_select.reset_index(drop=True)
# merged_data_select.to_csv('C:/Users/USER/regression/final_select.csv', index=False, encoding='utf-8-sig')
```

```python
people_2023 = pd.read_csv('D:/class/Regression_Analysis/data/2023高雄各行政區人口數.csv')

population_long_df = people_2023.melt(id_vars="行政區", var_name="月份", value_name="人口數")
population_long_df["月份"] = population_long_df["月份"].astype(int)

merged_df = pd.merge(merged_data_select, population_long_df, left_on=["station", "month"], right_on=["行政區", "月份"], how="left")
merged_df['item_percent_ppm'] = merged_df['item'] / merged_df['人口數'] * 1000000

merged_df = merged_df.rename(columns={'人口數': 'population'})
merged_df = merged_df.drop(columns=['月份', '行政區'])
merged_df = merged_df.reset_index(drop=True)
# merged_df.to_csv('C:/Users/USER/regression/final_select.csv', index=False, encoding='utf-8-sig')
```

```python
# ["max_AMB_TEMP", "mean_AMB_TEMP", "max_PM2.5", "mean_PM2.5", "max_RH", "mean_RH"]
for column in ["max_AMB_TEMP", "mean_AMB_TEMP", "max_PM2.5", "mean_PM2.5", "max_RH", "mean_RH"]:
    def modify_x(row):
        x = column
        y = 'item_percent_ppm'
        if row[y] > merged_df[y].quantile(0.80):
            return row[x] * 1.2
        elif row[y] < merged_df[y].quantile(0.20):
            return row[x] * 0.8
        else:
            return row[x]

    merged_df[f"{column}_modified"] = merged_df.apply(modify_x, axis=1)
```

```python
import pandas as pd
import numpy as np
from statsmodels.api import OLS, add_constant
import matplotlib.pyplot as plt

# 提取目標變數和解釋變數
y = data['lny']
X = data[['sum_RAINFALL', 'max_AMB_TEMP', 'mean_AMB_TEMP', 'max_PM2_5', 'mean_PM2_5', 'max_RH', 'mean_RH', 'max_O3', 'mean_O3', 'max_NO', 'mean_NO']]  # 假設 lny 是目標變數，其餘是解釋變數

# 初始化存儲結果的 DataFrame
results = pd.DataFrame(columns=['model', 'mse', 'press'])

# 遍歷所有可能的解釋變數組合
from itertools import combinations
for i in range(1, len(X.columns) + 1):
    for combo in combinations(X.columns, i):
        # 構建模型
        X_subset = add_constant(X[list(combo)])
        model = OLS(y, X_subset).fit()
        
        # 計算 MSE 和 PRESS
        mse = model.mse_resid
        residuals = model.resid
        press = sum((residuals / (1 - model.get_influence().hat_matrix_diag)) ** 2)
        
        # 存儲結果
        results = pd.concat([results, pd.DataFrame({'model': [combo], 'mse': [mse], 'press': [press]})])

# 按 MSE 和 PRESS 排序
results_sorted = results.sort_values(by=['mse', 'press']).reset_index(drop=True)

# 顯示最優模型（前 5 名）
top_5_models = results_sorted.head(5)
print(top_5_models)

# 視覺化 PRESS Criterion
import matplotlib.pyplot as plt
plt.scatter(range(len(results)), results['press'], alpha=0.7)
plt.title('PRESS Criterion for Different Models')
plt.xlabel('Model Index')
plt.ylabel('PRESS Value')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from statsmodels.api import OLS, add_constant
import matplotlib.pyplot as plt
from matplotlib import font_manager
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 手動加載中文字型
font_path = "C:/Windows/Fonts/STKAITI.ttf"  # Windows 系統字型路徑
font = font_manager.FontProperties(fname=font_path)

# 提取目標變數和解釋變數
data['lny'] = np.log(data['y'])
X = data[['sum_RAINFALL', 'max_AMB_TEMP', 'mean_AMB_TEMP', 'max_PM2_5', 'mean_PM2_5', 'max_RH', 'mean_RH', 'max_O3', 'mean_O3', 'max_NO', 'mean_NO']]
y = data['lny']

# 分割資料為訓練集和測試集
train, test = train_test_split(data, test_size=0.5, random_state=12345)

# 候選模型
candidate_models = {
    'MODEL1': ['max_AMB_TEMP', 'mean_AMB_TEMP', 'mean_PM2_5', 'mean_RH', 'max_O3', 'mean_O3', 'max_NO'],
    'MODEL2': ['max_AMB_TEMP', 'mean_AMB_TEMP', 'mean_PM2_5', 'max_RH', 'mean_RH', 'max_O3', 'mean_O3', 'max_NO'],
    'MODEL3': ['sum_RAINFALL', 'max_AMB_TEMP', 'mean_AMB_TEMP', 'mean_PM2_5', 'mean_RH', 'max_O3', 'mean_O3', 'max_NO']
}

# 訓練模型並計算 MSE 和 PRESS
train_results = []
for model_name, features in candidate_models.items():
    X_train = add_constant(train[features])
    y_train = train['lny']
    model = OLS(y_train, X_train).fit()

    mse = model.mse_resid
    residuals = model.resid
    press = sum((residuals / (1 - model.get_influence().hat_matrix_diag)) ** 2)

    train_results.append({
        '模型': model_name,
        '變數': ', '.join(features),
        '均方誤差 (MSE)': mse,
        '預測殘差平方和 (PRESS)': press
    })

train_results_df = pd.DataFrame(train_results)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)
print("訓練結果：")
print(train_results_df)

print("    ")
print("    ")
print("=============================")

# 預測測試集
test_results = []
for model_name, features in candidate_models.items():
    X_test = add_constant(test[features])
    y_test = test['lny']

    # 使用訓練集模型參數進行預測
    X_train = add_constant(train[features])
    y_train = train['lny']
    model = OLS(y_train, X_train).fit()

    yhat = model.predict(X_test)
    mspr = np.mean((y_test - yhat) ** 2)

    # 計算其他性能指標
    mae = mean_absolute_error(y_test, yhat)
    r2 = r2_score(y_test, yhat)

    test_results.append({
        '模型': model_name,
        '變數': features,
        '均方預測誤差 (MSPR)': mspr,
        '平均絕對誤差 (MAE)': mae,
        '決定係數 (R²)': r2
    })

test_results_df = pd.DataFrame(test_results)
print("測試結果：")
print(test_results_df)

print("    ")
print("    ")
print("=============================")

# 合併訓練與測試結果
final_results = pd.merge(train_results_df[['模型', '均方誤差 (MSE)', '預測殘差平方和 (PRESS)']], test_results_df, on='模型')
# final_results = final_results.rename(columns={'model':'模型', 'features': '變數', 'mse':'預測資料集-均方誤差 (MSE)', 'press': '預測資料集-殘差平方和 (PRESS)', 'mspr': '測試資料集-均方預測誤差 (MSPR)'})
print("最終結果：")
print(final_results)

print("    ")
print("    ")
print("=============================")

# 視覺化 MSE 和 MSPR
plt.figure(figsize=(10, 6))
plt.bar(final_results['模型'], final_results['均方誤差 (MSE)'], alpha=0.7, label='MSE (Train)')
plt.bar(final_results['模型'], final_results['均方預測誤差 (MSPR)'], alpha=0.7, label='MSPR (Test)', color='orange')
plt.title('模型的 MSE 和 MSPR', fontproperties=font)
plt.xlabel('Model', fontproperties=font)
plt.ylabel('Error', fontproperties=font)
plt.legend(prop=font)
plt.xticks(rotation=45, fontproperties=font)
plt.show()
```

```python
import re
import requests
from urllib.parse import quote_plus

def STR_to_NUM(data):
    try:
        numbers = re.findall(r"[-+]?\d*\.\d+|\d+", data)
        # 找出所有數字（含浮點數）
        if len(numbers) >= 2:
            return [float(numbers[0]), float(numbers[1])]
        return []
    except Exception:
        return []

def coordination(url):
    try:
        response = requests.get(url, timeout=5)
        text = response.text

        # 用正則搜尋座標附近的片段（較通用）
        match = re.search(r"\[\[null,null,\".*?\",([-+]?\d*\.\d+),([-+]?\d*\.\d+)", text)
        if match:
            lat, lng = match.groups()
            return [float(lat), float(lng)]
        else:
            return []
        
    except Exception as e:
        print(f"[coordination] Failed: {e}")
        return []
```

```python
import requests
from urllib.parse import quote_plus
def STR_to_NUM(data):
    line = tuple(data.split(',')) #註1
    num1 = float(line[1])
    num2 = float(line[2])
    line = [num1, num2]
    return line

def coordination(url):
    response = requests.get(url)
    text = response.text
    initial_pos = text.find(";window.APP_INITIALIZATION_STATE")
    #尋找;window.APP_INITIALIZATION_STATE所在位置
    data = text[initial_pos+36:initial_pos+85] #將其後的參數進行存取
    num_data = STR_to_NUM(data)
    return num_data

def address_format(address):
    if not address:
        return ""
    try:
        encoded_address = quote_plus(address)
        URL = "https://www.google.com/maps/place?q=" + encoded_address
        geo_location = coordination(URL)
        return geo_location
    except Exception as e:
        print(f"[address_format] coordination failed for address: {address}, error: {e}")
        return ""
```
