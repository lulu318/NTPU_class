#### <font color=#0000FF>step 1 : 讀入資料</font>

```python
uri = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data'
df = pd.read_csv(uri, header=None)
print(df.shape)

X, y = df.to_numpy()[:, 2:], df.to_numpy()[:, 1]    # y為字元型標籤
print(Counter(y))
```

#### <font color='gray'>(Optional) step 2 : 轉換類別型資料的 y 值為 0, 1, 2, ....</font>

```python
encoder = LabelEncoder()       # 使用LabelEncoder類將其轉換為0開始的數值型
y = encoder.fit_transform(y)   # .fit_transform(y) 等同於先後做 (1) .fit(y) (2) .transform(y)
#1. encoder.fit(y)                            
#2. encoder.transform(y)

print(Counter(y))
```

#### <font color=#0000FF>step 3 : 切割資料集</font>

```python
X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.2, random_state=None)
```

#### <font color=#0000FF>step 4 : 將 X 標準化</font>

```python
print('Before : \n', X_train)
print('shape of X_train : \n', X_train.shape)

scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)

print('After : \n', X_train)
print('shape of X_train : \n', X_train.shape)
```

#### <font color=#0000FF>step 5 : 使用 PCA 將 X_test 降維</font>

```python
print('Before : shape of X_train : {}'.format(X_train.shape))
pca = PCA(n_components=2)
X_train = pca.fit_transform(X_train)
X_test = pca.fit_transform(X_test)
print('After : shape of X_train : {}'.format(X_train.shape))
```

#### <font color=#0000FF>step 6 : 建模</font>

```python
lr = LogisticRegression(random_state=1)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

#### <font color=#0000FF>step 7 : 預測</font>

```python
lr.predict(X_test)
```

### <font color=#FF0000>將以上 step 4 ~ step 7 的過程打包成一個 pipeline

#### <font color=blue>使用 sklearn.pipeline.Pipeline 類別</font>

```python
# 他就是一個管道的概念
from sklearn.pipeline import Pipeline  # 建議用類別的部分
```

```python
%%time
# 1.讀取資料檔案，以及前置處理
uri = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data'
df = pd.read_csv(uri, header=None)
X, y = df.values[:, 2:], df.values[:, 1]    # y為字元型標籤

# encoder = LabelEncoder()
# y = encoder.fit_transform(y)   # (Optional)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)

#----------#----------#----------#----------#----------#----------#----------#----------#

# 2.建置 pipeline 
pipeline = Pipeline([("scaler", StandardScaler()),             # step 4
                     ("pca", PCA(n_components=2)),             # step 5
                     ("model", LogisticRegression())],)        # step 6

# 3.配適與正確率評估
pipeline.fit(X_train, y_train)
print(f"train acc : {pipeline.score(X_train, y_train):.3f}")
print(f"test acc : {pipeline.score(X_test, y_test):.3f}")

# 4.預測
pipeline.predict(X_test)
```

#### <font color=blue>細部探討 Pipeline 內的每一個物件，以及其method(s)</font>

```python
Pipeline([('sc', StandardScaler()), 
          ('pca', PCA(n_components=2)), 
          ('clf', LogisticRegression(random_state=1))])
```

```python
print("---【列出pipeline裡面的steps】----------------------")
print(pipeline)
print("---【取出pipeline裡面的特定物件】----------------------")
print(pipeline["scaler"])
print(pipeline["pca"])
print(pipeline["model"])
print("---【使用pipeline裡面個別的物件】----------------------")
print(type(pipeline["scaler"]))
pipeline["scaler"].fit_transform(X_test)
```

### <font color=#0000FF>堆疊多個 Pipelines，用以找到最佳的模型</font>

```python
%%time
pipeline_lr=Pipeline([('scaler'         , StandardScaler()),
                      ('pca'            , PCA(n_components=2)), 
                      ('estimator'      , LogisticRegression())])
pipeline_dt=Pipeline([('scaler'         , StandardScaler()),
                      ('pca'            , PCA(n_components=2)),
                      ('estimator'      , DecisionTreeClassifier())])
pipeline_rf = Pipeline([('scaler'       , StandardScaler()),
                        ('pca'          , PCA(n_components=2)),
                        ('estimator'    , RandomForestClassifier())])
pipeline_knn=Pipeline([('scaler'        , StandardScaler()),
                       ('pca'           , PCA(n_components=2)),
                       ('estimator'     , KNeighborsClassifier())])

pipelines = [pipeline_lr, 
             pipeline_dt, 
             pipeline_rf, 
             pipeline_knn]

#----------#----------#----------#----------#----------#----------#----------#----------#

for idx, pipeline in enumerate(pipelines, start=1) :
    print(f"【{idx}】{pipeline['estimator'].__class__.__name__}")
    print(pipeline.named_steps)
    pipeline.fit(X_train, y_train)
    print(f"train acc : {pipeline.score(X_train, y_train):.3%}")
    print(f"test acc : {pipeline.score(X_test, y_test):.3%}\n")
```
