1. ID number
2. Diagnosis (M = malignant, B = benign) : （M =惡性，B =良性）

3-32 , Ten real-valued features are computed for each cell nucleus:

3. radius (mean of distances from center to points on the perimeter) : 半徑（從中心到外圍點的距離的平均值）
4. texture (standard deviation of gray-scale values) : 紋理（灰度值的標準偏差）
5. perimeter : 周長
6. area : 面積
7. smoothness (local variation in radius lengths) : 平滑度（半徑長度的局部變化）
8. compactness (perimeter^2 / area - 1.0) : 密實度（周長^ 2 /面積-1.0）
9. concavity (severity of concave portions of the contour) : 凹度（輪廓凹部的嚴重程度）
10. concave points (number of concave portions of the contour) : 凹點（輪廓的凹入部分的數量）
11. symmetry : 對稱
12. fractal dimension ("coastline approximation" - 1) : 分形維數（“海岸線近似”-1）

### <font color=#0000FF>sklearn.model_selection.GridSearchCV</font>

#### <font color='red'>【以下scoring設定的參考】</font>
- [3.3. Metrics and scoring: quantifying the quality of predictions](https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter) <font color=red></font>

```python
from sklearn.model_selection import GridSearchCV
```

```python
%%time
iris = pd.read_csv(r"C:\Data\PyMLData\iris.csv", 
                   sep=',', encoding='cp950', engine='python')

X = iris.iloc[:,:4].to_numpy()
y = iris.iloc[:, 4].to_numpy()

X_train, X_test, y_train ,y_test = \
        train_test_split(X, y, test_size=0.2, random_state=None)
X
```

```python
%%time
# 設定管道/ 流程. List of (name of step, estimator) tuples
pipeline = Pipeline([("scaler", StandardScaler()),
                     ("pca", PCA()),
                     ("rf", RandomForestClassifier())])
print(pipeline.named_steps)  # 顯示管道步驟.

# 你希望的所有超參數值
param_grid = {"rf":[RandomForestClassifier()],
              "rf__n_estimators":[10, 50], # 命名規則: rf__該函數真實之參數名稱 (須一致)
              "rf__criterion":["gini", "entropy"],
              "rf__max_samples":[30, 40, 50],
              "pca":[PCA()],
              "pca__n_components":[2, 3, 4]  # 根據資料集的變數個數 設定
}
# 計180 組 = 2 * 2 * 3 * 3 * 5
# 最後的 5 是 cv=5

searchCV = GridSearchCV(estimator=pipeline,
                        param_grid=param_grid,    # 給定希望的 超參數
                        scoring="accuracy", 
                        n_jobs=-1, 
                        refit=True,
                        cv=5)
best_model = searchCV.fit(X_train, y_train)
print(f"train acc : {best_model.score(X_train, y_train)}")
print(f"test acc : {best_model.score(X_test, y_test)}")
```

```python
# 輸出最佳模型之超參數選擇.
print(best_model.best_estimator_)
print(best_model.best_params_)
print(best_model.best_score_)
```

### <font color=#0000FF>sklearn.model_selection.RandomizedSearchCV</font>

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
%%time
iris = pd.read_csv(r"C:\Data\PyMLData\iris.csv", 
                   sep=',', encoding='cp950', engine='python')

X = iris.iloc[:,:4].to_numpy()
y = iris.iloc[:, 4].to_numpy()

X_train, X_test, y_train ,y_test = \
        train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
%%time
# 設定管道/ 流程. List of (name of step, estimator) tuples
pipeline = Pipeline([("scaler", StandardScaler()),
                     ("pca", PCA()),
                     ("rf", RandomForestClassifier())])
print(pipeline.named_steps)  # 顯示管道步驟.

# 你希望的所有超參數值
param_grid = {"rf":[RandomForestClassifier()],
              "rf__n_estimators":[10, 50], # 命名規則: rf__該函數真實之參數名稱 (須一致)
              "rf__criterion":["gini", "entropy"],
              "rf__max_samples":[30, 40, 50],
              "pca":[PCA()],
              "pca__n_components":[2, 3, 4]  # 根據資料集的變數個數 設定
}
# 計180 組 = 2 * 2 * 3 * 3 * 5
# 最後的 5 是 cv=5

searchCV = RandomizedSearchCV(estimator=pipeline,
                        param_distributions=param_grid,    # 給定希望的 超參數
                        scoring="accuracy", 
                        n_jobs=-1, 
                        refit=True,
                        cv=5)
best_model = searchCV.fit(X_train, y_train)
print(f"train acc : {best_model.score(X_train, y_train)}")
print(f"test acc : {best_model.score(X_test, y_test)}")
```

```python
print(best_model.best_estimator_)
print(best_model.best_params_)
print(best_model.best_score_)
```
