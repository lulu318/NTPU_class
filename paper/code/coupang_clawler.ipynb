```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from bs4 import BeautifulSoup
import time
import urllib.parse

# 設定 Chrome 瀏覽器的選項
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36")
prefs = {"profile.default_content_setting_values.notifications": 2}
chrome_options.add_experimental_option("prefs", prefs)

# 你的商品關鍵字
keyword = "可口可樂"
# 設定 Firefox 瀏覽器的選項
firefox_options = FirefoxOptions()
firefox_options.add_argument("--headless") # 啟用無頭模式

# 初始化 Firefox 驅動
# 確保 geckodriver.exe 檔案在程式碼同一個資料夾或已設定環境變數
driver = webdriver.Firefox(options=firefox_options)

encoded_keyword = urllib.parse.quote(keyword)
url = f"https://www.tw.coupang.com/search?component=&q={encoded_keyword}&channel=user"

try:
    print(f"正在載入酷澎網頁：{url}")
    driver.get(url)

    # 等待商品列表中的某個元素出現，以確保頁面已經載入完成
    # 這裡假設商品列表的容器是 <li class="search-product-list-item">
    wait = WebDriverWait(driver, 30)
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'li.search-product-list-item')))
    
    print("網頁載入完成，正在獲取頁面原始碼...")
    page_source = driver.page_source
    
except Exception as e:
    print(f"載入網頁或等待元素時發生錯誤：{e}")
finally:
    driver.quit() # 結束後務必關閉瀏覽器

print("瀏覽器已關閉。開始使用 BeautifulSoup 解析資料...")
```
