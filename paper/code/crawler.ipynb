```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium_stealth import stealth
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import random
import json

# ✅ 設定 Chrome 非 headless 模式
options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--start-maximized")  # 全螢幕開啟

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ✅ 使用 Selenium Stealth 隱藏自動化痕跡
stealth(driver,
        languages=["zh-TW", "zh"],
        vendor="Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=True)

KEYWORD = "耳機"
URL = f"https://shopee.tw/search?keyword={KEYWORD}"
driver.get(URL)

input("🛑 請手動登入 Shopee，完成後按 Enter 繼續...")

# ✅ 隨機等待，模擬人類行為
time.sleep(random.uniform(3, 6))

# ✅ 模擬滑動頁面
for i in range(10):
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(random.uniform(1, 2))

# ✅ 擷取商品資料
soup = BeautifulSoup(driver.page_source, "html.parser")
items = soup.select("div.shopee-search-item-result__item")

results = []
for item in items:
    name_tag = item.select_one("div[data-sqe='name']")
    price_tag = item.select_one("div[data-sqe='price']")
    img_tag = item.select_one("img")

    if name_tag and price_tag and img_tag:
        results.append({
            "name": name_tag.text.strip(),
            "price": price_tag.text.strip(),
            "image": img_tag["src"]
        })

driver.quit()

with open("shopee_products_stealth.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"✅ 擷取完成，共 {len(results)} 筆商品")
```
