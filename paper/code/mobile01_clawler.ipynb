```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

keyword = "情緒勒索"

# 設定瀏覽器選項
options = Options()
# options.add_argument('--headless')  # 不開啟視窗
options.add_argument('--disable-blink-features=AutomationControlled')  # 防偵測
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120 Safari/537.36")

# 啟動瀏覽器
driver = webdriver.Chrome(options=options)

# 目標文章（你可以換成任何 Mobile01 討論串）
url = f"https://www.mobile01.com/googlesearch.php?query={keyword}"
driver.get(url)
time.sleep(2)  # 等待 JS 載入

# # 抓取標題
# try:
#     title = driver.find_element("css selector", "h1").text
#     print(f"✅ 成功進入頁面！標題是：{title}")
# except Exception as e:
#     print("❌ 無法抓取標題，可能被封鎖或網頁結構變動")

# driver.quit()

# 不切 iframe，直接抓搜尋結果
results = driver.find_elements(By.CSS_SELECTOR, ".gs-title a")

# 用 set 去重
seen_links = set()
unique_articles = []

for r in results:
    title = r.text.strip()
    link = r.get_attribute("href")

    if link and "topicdetail.php" in link and link not in seen_links:
        seen_links.add(link)
        unique_articles.append((title, link))

driver.quit()

# 顯示結果
print(f"✅ 共抓到 {len(unique_articles)} 篇不重複文章")
for title, link in unique_articles:
    print(f"標題：{title if title else '[無標題]'}\n連結：{link}\n")
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# for title, link in unique_articles[0]:
title, link = unique_articles[0]
driver = webdriver.Chrome(options=options)
driver.get(link)
time.sleep(2)

try:
    WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "h1.t2"))
        )
    page_title = driver.find_element(By.CSS_SELECTOR, "h1.t2").text.strip()
    # # 抓作者
    # try:
    #     WebDriverWait(driver, 10).until(
    #         EC.presence_of_element_located((By.CSS_SELECTOR, ".c-authorInfo_id a"))
    #     )
    #     author = driver.find_element(By.CSS_SELECTOR, ".c-authorInfo_id a").text.strip()
    # except:
    #     author = "[無作者]"

    # 抓時間
    try:
        post_time = driver.find_element(By.CSS_SELECTOR, ".o-fNotes.o-fSubMini").text.strip()
    except:
        post_time = "[無時間]"

    # 抓主文內容
    content = driver.find_element(By.CSS_SELECTOR, "div[itemprop='articleBody']").text.strip()

    print(f"\n標題：{page_title}")
    # print(f"作者：{author}")
    print(f"時間：{post_time}")
    print(f"內容預覽：{content[:100]}...")  # 印前 200 字

    # 抓所有回文（排除主文）
    all_articles = driver.find_elements(By.CSS_SELECTOR, "article.c-articleLimit")
    reply_texts = []
    for a in all_articles[1:]:  # 跳過第一篇主文
        reply = a.text.strip()
        # 過濾引用段落（如「自 XXX 引用」）
        if reply.startswith("自") and "引用" in reply:
            continue
        reply_texts.append(reply)
    print(f"[ 共抓到 {len(reply_texts)} 則留言 ]")
    for i, r in enumerate(reply_texts, 1):
        print(f" > 留言 {i}：{r[:100]}")
except Exception as e:
    print(f"⚠️ 解析失敗：{link}")
driver.quit()
```
