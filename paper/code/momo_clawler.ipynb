```python
import requests
from urllib.parse import quote
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36',
    }  
keyword = "可口可樂" 
url = f"https://www.momoshop.com.tw/search/searchShop.jsp?keyword={quote(keyword)}&isFuzzy=0&searchType=1"
res = requests.get(url, headers=headers)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.title.text)
print(soup.title)
```

```python
import json
import re, json, html
# soup = BeautifulSoup(html_text, "html.parser")
# 假設 html_text 是你抓到的 HTML 原始碼

# 找出第一個 type 為 application/ld+json 的 <script> 標籤
script_tag = soup.find("script", {"type": "application/ld+json"})

# 取得 JSON 字串並清理空白
json_text = script_tag.get_text(strip=True)
cleaned = re.sub(r',\s*([\]}])', r'\1', json_text)
# 將 JSON 字串轉成 Python 字典
try:
    data = json.loads(cleaned)
except json.JSONDecodeError as e:
    print("JSON 解析失敗:", e)
    data = None

# 印出結果確認
print(json.dumps(data, indent=2, ensure_ascii=False))
```

```python
import pandas as pd
pd_momo = pd.DataFrame(data['mainEntity']['itemListElement'])
pd_momo
```

```python
import requests
import json
import time
import random

url = "https://eccapi.momoshop.com.tw/user/getGoodsCommentList"
headers = {
    "Content-Type": "application/json;charset=UTF-8",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Origin": "https://www.momoshop.com.tw",
    "Referer": "https://www.momoshop.com.tw/"
    # 可加入 Cookie 模擬登入狀態（視情況而定）
}

payload = {
    "host": "web",
    "goodsCode": "10514714",
    "curPage": 1,
    "filterType": "total",
    "multiFilterType": ["hasComment"],
    "custNo": ""
}

response = requests.post(url, headers=headers, json=payload)
data = response.json()
```
