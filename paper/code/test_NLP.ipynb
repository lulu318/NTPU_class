```python
from snownlp import SnowNLP
import time
# 自然語言分析範例

print('[SnowNLP result]')
start_time = time.time()
input = '今天天氣很好'
print(f'\nINPUT 字串： {input}')
print('分數： ' + str(SnowNLP(SnowNLP(input).sentences[0]).sentiments))
print(f"   > 花費時間{(time.time() - start_time):.6f} seconds")

start_time = time.time()
input = '我很生氣'
print(f'\nINPUT 字串： {input}')
print('分數： ' + str(SnowNLP(SnowNLP(input).sentences[0]).sentiments))
print(f"   > 花費時間{(time.time() - start_time):.6f} seconds")
```
